```python
import pandas as pd
import numpy as np

file_path = '/content/BRT  Non BRT Route Details  Bus Stop LatLong (1).xls'

xls = pd.ExcelFile(file_path)
sheet_names = xls.sheet_names

for sheet_name in sheet_names:

    df = pd.read_excel(xls, sheet_name)


    if "Stop Name Marathi" in df.columns:
        df.drop(columns=["Stop Name Marathi"], inplace=True)

    if "Route Description Marathi" in df.columns:
      df.drop(columns=["Route Description Marathi"], inplace=True)

    df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

    output_csv_file_path = f'/content/{sheet_name}.csv'

    df.to_csv(output_csv_file_path, index=False)
```

```python
import pandas as pd

route_df = pd.read_csv('filtered_bus_stops.csv')
bus_stops_df = pd.read_csv('filtered_file.csv')

route_df['Stop Name'] = route_df['Stop Name'].str.lower().str.strip()
bus_stops_df['StopName'] = bus_stops_df['StopName'].str.lower().str.strip()
route_df['Direction'] = route_df['Direction'].str.upper().str.strip()
bus_stops_df['Direction'] = bus_stops_df['Direction'].str.upper().str.strip()

routes_per_stop = route_df.groupby(['Stop Name', 'Direction'])['Route'].apply(list).reset_index()
routes_per_stop['Route'] = routes_per_stop['Route'].apply(lambda x: list(set(x)))

bus_stops_with_routes = bus_stops_df.merge(routes_per_stop, left_on=['StopName', 'Direction'], right_on=['Stop Name', 'Direction'], how='left')

bus_stops_with_routes.drop('Stop Name', axis=1, inplace=True)
bus_stops_with_routes.rename(columns={'Route': 'Routes'}, inplace=True)

bus_stops_with_routes.to_csv('updated_Bus_stop.csv', index=False)
bus_stops_with_routes
```

```python
import pandas as pd

bus_stops_df = pd.read_csv('updated_Bus_stop.csv')

bus_stops_df['Latitude'] = pd.to_numeric(bus_stops_df['Latitude'], errors='coerce')
bus_stops_df['Longitude'] = pd.to_numeric(bus_stops_df['Longitude'], errors='coerce')

bus_stops_df.dropna(subset=['Latitude', 'Longitude'], inplace=True)

median_latitude = bus_stops_df['Latitude'].median()
median_longitude = bus_stops_df['Longitude'].median()

print(f"Median Latitude: {median_latitude}, Median Longitude: {median_longitude}")

def determine_zone(lat, lon, median_lat, median_lon):
    if lat < median_lat:
        if lon < median_lon:
            return 'Zone 1'
        else:
            return 'Zone 2'
    else:
        if lon < median_lon:
            return 'Zone 4'
        else:
            return 'Zone 3'

bus_stops_df['Zone'] = bus_stops_df.apply(
    lambda row: determine_zone(row['Latitude'], row['Longitude'], median_latitude, median_longitude), axis=1
)

print(bus_stops_df['Zone'].value_counts())
new_order = ['ID', 'Direction', 'StopName','StopCode','Latitude', 'Longitude', 'Status','StopType', 'Routes', 'Zone']
bus_stops_df = bus_stops_df[new_order]
bus_stops_df.to_csv('BusStops.csv', index=False)
bus_stops_df
```
