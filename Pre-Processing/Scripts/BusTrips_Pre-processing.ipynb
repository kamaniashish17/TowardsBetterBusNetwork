```python
import pandas as pd

file_path = '/Users/tanayalad/Desktop/Tanaya Files/Masters/Classes/2nd Semester/CSE 578 - DV/DV Project/Cleaned Data/BusRoutes.csv'

bus_routes_data = pd.read_csv(file_path)

filtered_columns = [col for col in bus_routes_data.columns if not col.startswith('Unnamed:')]
bus_routes_data = bus_routes_data[filtered_columns]

bus_routes_data
```

```python
bus_routes_values = bus_routes_data.values
bus_routes_values
```

```python
from datetime import time, timedelta
import numpy as np

tripId = 0

bus_trips = []

for val in bus_routes_values:
    for t in range(6, 18):
        row = []
        tripId += 1
        row.append(tripId)
        row.append(val[1])
        row.append(val[13])
        row.append(val[12])
        row.append(time(hour = t, minute = 0))
        row.append(val[4])
        num_val_in = eval(val[2*(t+1)])
        avg_check_in = sum(num_val_in)/len(num_val_in)
        row.append(avg_check_in)
        num_val_out = eval(val[(2*(t+1))+1])
        avg_check_out = sum(num_val_out)/len(num_val_out)
        row.append(avg_check_out)
        avg_load = np.random.randint(1, 70)
        row.append(avg_load)
        bus_trips.append(row)
        
print(bus_trips)
```

```python
import pandas as pd

column_names = ['Trip ID', 'Route ID', 'Service Cost', 'Construction Cost', 'Timestamp', 'Average Speed', 'Average Check In', 'Average Check Out', 'Average Load']

df = pd.DataFrame(bus_trips, columns=column_names)

df.head()
```

```python
df.to_csv('BusTrips.csv', index=False)
```
